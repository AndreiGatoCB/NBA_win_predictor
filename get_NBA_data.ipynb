```python
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import lxml.html
import time
```

```python
SEASONS_1 = list(range(2000, 2005))
SEASONS_2 = list(range(2005, 2010))
SEASONS_3 = list(range(2010, 2015))
SEASONS_4 = list(range(2015, 2020))
SEASONS_5 = list(range(2020, 2025))
```

```python
DATA_DIR = 'NBA_data'
STANDINGS_DIR = os.path.join(DATA_DIR, 'standings')
SCORES_DIR = os.path.join(DATA_DIR, 'scores')
```

```python
def get_html(url, label, clase, id):
    page = requests.get(url)
    soup = bs(page.content, 'html.parser')
    title = soup.find('title').text
    print(title)
    if label == '':
        html = soup.find_all(id=id)

    else:
        html = soup.find_all(label, class_=clase)

    return str(html), title
```

```python
def scrape_season(season):
    url = f'https://www.basketball-reference.com/leagues/NBA_{season}_games.html'
    months = get_html(url, 'div', 'filter')
    soup = bs(months, 'lxml')
    months_links = soup.find_all('a', href=True)

    href = [i['href'] for i in months_links]
    # print(href)
    standing_pages = [f'https://www.basketball-reference.com{l}' for l in href]
    # print(standing_pages)
    title = ''
    for url in standing_pages:
        save_path = os.path.join(STANDINGS_DIR, url.split('/')[-1])
        # print(save_path)
        if os.path.exists(save_path):
            continue
        html, title = get_html(url, '', 'all_schedule', 'all_schedule')
        # print(html)
        with open(save_path, 'w+') as f:
            f.write(html)

    return title
```

```python
standings_file = os.path.join(STANDINGS_DIR, standings_files[1])
# print(standings_file)
```

```python
def scrape_game(standings_file):
    with open(standings_file, 'r') as f:
        html = f.read()

    soup = bs(html)
    links = soup.find_all('a')
    hrefs = [l.get('href') for l in links]
    box_scores = [l for l in hrefs if l and 'boxscore' in l and '.html' in l]
    box_scores = [f'https://www.basketball-reference.com{l}' for l in box_scores]

    title = ''
    for url in box_scores:

        save_path = os.path.join(SCORES_DIR, url.split('/')[-1])
        if os.path.exists(save_path):
            continue

        html, title = get_html(url, '', 'box', 'content')

        if not html:
            continue
        with open(save_path, 'w+', encoding='utf-8') as f:
            f.write(html)
            time.sleep(0.5)

    return title
```

```python
i = 0
for f in standings_files:
    file_path = os.path.join(STANDINGS_DIR, f)
    title = scrape_game(file_path)

    if not 'error' in title:
        print(i)
        i += 1
    else:
        print('error')
        break

```
